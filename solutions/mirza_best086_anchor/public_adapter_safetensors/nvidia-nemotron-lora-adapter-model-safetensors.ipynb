## Create Submission ZIP File

The following code packages the required adapter files into a ZIP archive for submission.

```python
import os, zipfile, json

ADAPTER_DIR = '/kaggle/input/datasets/assiabenazzouz/adappter-v32-epoch-5/adapter'
zip_path = '/kaggle/working/submission.zip'

with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as zf:
    for fname in ["adapter_config.json", "adapter_model.safetensors"]:
        fpath = os.path.join(ADAPTER_DIR, fname)
        if os.path.exists(fpath):
            zf.write(fpath, fname)
            print(f"✓ Added: {fname}")
        else:
            print(f"× Missing: {fname}")

size = os.path.getsize(zip_path)/1024/1024
print(f"submission.zip: {size:.1f} MB")
```

### Explanation
1. Define the directory containing the adapter files.
2. Specify the output ZIP file location.
3. Create a compressed ZIP archive using `zipfile.ZipFile`.
4. Check whether the required files exist:
   - `adapter_config.json`
   - `adapter_model.safetensors`
5. Add available files to the ZIP archive.
6. Display missing files (if any).
7. Calculate and display the final ZIP file size.

### Output Example

```text
✓ Added: adapter_config.json
✓ Added: adapter_model.safetensors
submission.zip: 125.4 MB
```

### Purpose

This script is commonly used in Kaggle competitions or model deployment workflows where only the adapter weights and configuration files need to be submitted. It ensures that the required files are packaged into a single `submission.zip` file ready for upload.

````python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("assiabenazzouz/adappter-v32-epoch-5")

print("Path to dataset files:", path)
````

````python
import os, zipfile, json

ADAPTER_DIR = '/kaggle/input/datasets/assiabenazzouz/adappter-v32-epoch-5/adapter'
zip_path = '/kaggle/working/submission.zip'

with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as zf:
    for fname in ["adapter_config.json", "adapter_model.safetensors"]:
        fpath = os.path.join(ADAPTER_DIR, fname)
        if os.path.exists(fpath):
            zf.write(fpath, fname)
            print(f"✓ Added: {fname}")
        else:
            print(f"× Missing: {fname}")

size = os.path.getsize(zip_path)/1024/1024
print(f"submission.zip: {size:.1f} MB")
````
